## Configuracion

**De linea:**
- linestyle = tipo
  - "-" o "solid": linea solida
  - "--" o "dashed": linea discontinua
  - "-." o "dashot": guiones y puntos
  - ":" o "dotted": linea de puntos
  - "none", "o": sin linea
- linewidth = ancho
- color = color de línea
  - "blue", "green", "red"...
  - "b", "o", "g"
  - (r, g, b): valores entre 0 y 1
  - "#RRGGBB"

**Marcas**
- marker = marcador
  - "o": círculo
  - "*": estrella
  - ".": punto

```python
# Actividad ejemplo 1: Graficar puntos con ruido
# Generar un conjunto de 50 elementos Lineales del 0 al 10, genera a partir de ellos una relación Linea para crear
# asi una Linea en la gráfica. Al final, agrégales ruido aleatorio
# Luego graficalos con plt.plot()

import numpy as np


# Datos base: vamos a generar esos 50 puntos entre el 0 y el 10
x = np.linspace(0, 10, 50)
y = 2 * x + 3 # Relacion lineal

# Agregar ruido aleatorio a apartir de una distribucion normal
y_noisy = y + np.random.normal(0, 2, size = len(x))

plt.figure(figsize=(8,5))
plt.plot(x, y_noisy, label="Datos con ruido", linestyle = "-", marker = "o")
plt.plot(x, y, label = "Línea original (sin ruido)", color = "red", linestyle ="--")
plt.title("Puntos con ruido")
plt.xlabel("Eje X")
plt.ylabel("Eje Y (con ruido)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Actividad ejemplo2: comparar funciones matematicas
# Generar un grafico con varias funciones matematicas (seno, coseno, exponencial)

# Datos base: genera 100 numeros desde -2pi hasta 2pi
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
sin_x = np.sin(x)
cos_x = np.cos(x)
exp_x = np.exp(x/4)

plt.figure(figsize=(8, 5))
plt.plot(x, sin_x, label='sen', color='blue')
plt.plot(x, cos_x, label='cos', color='green')
plt.plot(x, exp_x, label='exp', color='red')

plt.axhline(0, c="black", linestyle = "--", linewidth = 2)
plt.axvline(0, c="black", linestyle = "--", linewidth = 2)

plt.title('Comparación de Funciones Matemáticas')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Clasificacion basada en el signo del producto

# Generar unos cuantos numeros
n_point = 200
x = np.random.uniform(-5, 5, n_point)
y = np.random.uniform(-5, 5, n_point)

group = x*y > 0 # Aqui los clasificamos

# Grafica
plt.figure(figsize=(8, 8))

plt.scatter(x[group], y[group], color = "blue", label = "Grupo 1 (+)")
plt.scatter(x[~group], y[~group], color = "red", label = "Grupo 2 (-)")

plt.axhline(0, c="black", linestyle = "--", linewidth = 2)
plt.axvline(0, c="black", linestyle = "--", linewidth = 2)

plt.title('Clasificacion basada en el signo del producto')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from matplotlib.figure import figaspect
# Actividd de adelanto de histograma: Subplots para comparar diferentes distribuciones
# Crear subgraficos que muestren histogramas de distribucion distintas (normla, uniforme y binomial)

# Grenerar datos
normal_data = np.random.normal(0, 1, 1000)
uniform_data = np.random.uniform(-3, 3, 1000)
binomial_data = np.random.binomial(10, 0.5, 1000)

# Crear subgrafico
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey = True)
axes[0].hist(normal_data, bins = 20, color = "blue", alpha = 0.7)
axes[0].set_title("Distribucion normal")

axes[1].hist(uniform_data, bins = 20, color = "green", alpha = 0.7)
axes[1].set_title("Distribucion uniforme")

axes[2].hist(binomial_data, bins = 20, color = "orange", alpha = 0.7)
axes[2].set_title("Distribucion binomial")

plt.suptitle("Comparacion de distribuciones")
plt.tight_layout()
plt.show()
```

## Figuras multigraficas

plt.subplot(nrows, ncols, index)
- nrows = numero de filas de la cuadrilla
- ncols = numero de columnas de la cuadrilla
- index = la posicion del subgrafico donde dibuja (enumerado de izquierda a derecha y de arriba abajo, empezando en 1)

```python
x = (3,2,1)
y = (1,2,3)

plt.subplot(2,3,1)
plt.plot(y)

plt.subplot(2,3,2)
plt.plot(x)

plt.subplot(2,3,3)
plt.plot(x,y)

plt.show()
```

```python
x = np.linspace(0,10,100)
plt.figure(figsize=(10,6))

# 1º subgrafico
plt.subplot(2,2,1)
plt.plot(x, np.sin(x))
plt.title("seno")

# 2º subgrafico
plt.subplot(2,2,2)
plt.plot(x, np.cos(x), color="red")
plt.title("coseno")

# 3º subgrafico
plt.subplot(2,2,3)
plt.plot(x, np.tan(x), color="green")
plt.title("tangente")

# 4º subgrafico
plt.subplot(2,2,4)
plt.plot(x, np.exp(x), color="yellow")
plt.title("exponencial")

plt.tight_layout()
plt.show()
```

```python
# Bonus: notacion compacta
plt.subplot(221)
plt.plot(x, np.sin(x))
plt.title("seno")
plt.show()
```
