9. El juego de las n-Torresconsiste en colocar en un tablero de ajedrez, de nxn casillas, a n torres
de forma que ninguna pueda capturar a otra. Cada fila 0 columna de un tablero de ajedrez se
comienza a numerar desde el 1. Mostrar gráficamente la posicion de las torres, para mayor comodidad anadir a la grafica una rejilla.

```python
def dibujar_n_torres_matplotlib(n):

    tablero = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            if (i + j) % 2 == 0:
                tablero[i, j] = 1

    fig, ax = plt.subplots(figsize=(8, 8))
    ax.imshow(tablero, cmap='Pastel1', origin='upper')

    for i in range(n):
        ax.text(i, i, '♜', va='center', ha='center', fontsize=30, color='black')

    ax.set_xticks(np.arange(n))
    ax.set_yticks(np.arange(n))
    ax.set_xticklabels(np.arange(1, n + 1))
    ax.set_yticklabels(np.arange(1, n + 1))

    ax.set_xticks(np.arange(-0.5, n, 1), minor=True)
    ax.set_yticks(np.arange(-0.5, n, 1), minor=True)
    ax.grid(which='minor', color='gray', linestyle='-', linewidth=1.5)

    ax.set_title(f"Problema de las {n}-Torres", fontsize=16, pad=20)
    plt.tick_params(axis='both', which='major', labelsize=12)

    plt.show()

dibujar_n_torres_matplotlib(8)
```

16. En el juego piedra-papel-tijeras v2.0, un grupo numeroso de personas se distribuyen aleatoriamente en tres equipos: el equipo piedra, el equipo papel y el equipo tijeras. Cada jugador
tira un dado de tres caras y con este método se decide de qué equipo forma parte.
El problema es que hay veces que los equipos quedan muy desigualados, por ejemplo 180
personas son piedra, 43 son papel y solo 4 son tijeras. Un equipo se considera en ventaja si
contiene más de 10 jugadores que el resto. En este caso, se realiza un nuevo sorteo.
Mostrar de forma interactiva el proceso de construccién de equipos, haciendo que el resultado sea apreciable de forma gráfica.

```python
def simulacion_equipos_final(n_personas):
    historial_sorteos = []
    equilibrado = False

    while not equilibrado:

        tiradas = np.random.randint(0, 3, size=n_personas)
        conteo = [np.sum(tiradas == 0), np.sum(tiradas == 1), np.sum(tiradas == 2)]

        diferencia = max(conteo) - min(conteo)

        historial_sorteos.append({'datos': conteo,'dif': diferencia,'exito': diferencia <= 10})

        if diferencia <= 10:
            equilibrado = True

    n_intentos = len(historial_sorteos)

    indices_a_mostrar = list(range(n_intentos))

    fig, axes = plt.subplots(1, len(indices_a_mostrar), figsize=(18, 5), sharey=True)

    # Si solo hay un gráfico
    if n_intentos == 1: axes = [axes]

    nombres = ['Piedra', 'Papel', 'Tijeras']

    for i, idx in enumerate(indices_a_mostrar):
        sorteo = historial_sorteos[idx]
        ax = axes[i]

        color_barra = 'green' if sorteo['exito'] else 'red'

        barras = ax.bar(nombres, sorteo['datos'], color=color_barra, alpha=0.6, edgecolor='black')
        ax.bar_label(barras, padding=3, fontweight='bold')

        titulo = f"Sorteo #{idx + 1}\nDif: {sorteo['dif']}"
        if sorteo['exito']: titulo += " (¡OK!)"
        ax.set_title(titulo, fontsize=12, color='darkgreen' if sorteo['exito'] else 'darkred')

        ax.grid(axis='y', linestyle='--', alpha=0.5)
        ax.set_ylim(0, max(sorteo['datos']) + 20)

    plt.suptitle(f"Evolución del Sorteo para {n_personas} personas\n(Total de intentos: {n_intentos})", fontsize=16, y=1.05)
    plt.tight_layout()
    plt.show()

simulacion_equipos_final(300)
```

```python
n, m = 50, 50
superficie = np.full((n, m), 255, dtype=np.int16)

num_gotas = 5000
intensidad_gota = 25

for _ in range(num_gotas):

    fila = np.random.randint(0, n)
    columna = np.random.randint(0, m)

    superficie[fila, columna] -= intensidad_gota

superficie = np.clip(superficie, 0, 255)

plt.figure(figsize=(6, 6))
plt.imshow(superficie, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.savefig('humedad_lluvia.png', bbox_inches='tight')
print("Imagen guardada como 'humedad_lluvia.png'")

plt.show()
```

21. Un fractal es un objeto geométrico cuya estructura basica se repite a diferentes escalas. Estan
muy presentes en la naturaleza.
Los fractales pueden reproducirse como una funcién compleja iterativa, que suele tener la
forma:
Z=func(Z)+C
donde Z es una matriz bidimensional de números complejos (habitualmente todos sus elementos son 0+0j, y C es una matriz bidimensional constante de números complejos. Existen
muchas funciones, cada una de ellas, junto con C, determinan la forma del fractal.
Podemos crear la matriz C, con el codigo:
coord_x = np.arange(-1, 1, 2/600)
coord_y = np.arange(-1, 1, 2/600)
X, Y = np.meshgrid(coord_x, coord.y)
C = X + Y*1j matriz compleja constante de 600x600
Inicialmente Z serd una matriz compleja de valor 0.
Y podemos usar muchas funciones, por ejemplo:
Z=cos(Z/C)
Finalmente, tenemos que convertir la matriz compleja en una matriz real, de la siguiente
‘manera:
imagen=$$e^{-|Z|}$$
Escribir una matriz que tome los valores de un fractal. Mostrar el resultado como una imagen.

```python
import numpy as np
import matplotlib.pyplot as plt

coord_x = np.arange(-1, 1, 2/600)
coord_y = np.arange(-1, 1, 2/600)
X, Y = np.meshgrid(coord_x, coord_y)

C = X + Y*1j
Z = np.zeros_like(C, dtype=complex)

iteraciones = 3
for _ in range(iteraciones):

    Z = np.cos(Z / C)

# imagen = e^(-|Z|)
imagen = np.exp(-np.abs(Z))

plt.figure(figsize=(8, 8))
plt.imshow(imagen, cmap='magma')
plt.axis('off')
plt.show()
```
