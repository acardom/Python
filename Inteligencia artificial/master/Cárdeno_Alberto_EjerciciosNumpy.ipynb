6. En un comercio cada vez que se vende un producto se anota en una libreta el importe de venta. Al finalizar el día, se calculan la comisiones para cada producto:
- El 21% es el IVA.
- Del importe restante, el 10% es el beneficio obtenido.
- Y el resto, es el dinero que se destina para comprar nuevos productos.

  Calcular para cada producto los importes correspondientes y los totales.

````python
# Datos de entrada
importes_venta = np.array([121.00, 50.00, 242.00, 80.00])
FACTOR_IVA_TOTAL = 1.21
FACTOR_BENEFICIO = 0.10
FACTOR_REPOSICION = 0.90
NUM_PRODUCTOS = len(importes_venta)

# Importe Base (sin IVA)
importe_base = importes_venta / FACTOR_IVA_TOTAL

# Monto de IVA
monto_iva = importes_venta - importe_base

# Monto de Beneficio (10% del Base)
monto_beneficio = importe_base * FACTOR_BENEFICIO

# Monto de Reposición (90% del Base)
monto_reposicion = importe_base * FACTOR_REPOSICION


total_venta = np.sum(importes_venta)
total_iva = np.sum(monto_iva)
total_beneficio = np.sum(monto_beneficio)
total_reposicion = np.sum(monto_reposicion)


print("---------------------------------------------------------------------------------------")
print("| Prod. | Venta Total | Importe Base | IVA (21%) | Beneficio (10%) | Reposición (90%) |")
print("---------------------------------------------------------------------------------------")


for i in range(NUM_PRODUCTOS):
    print(f"| {i+1:5} | {importes_venta[i]:11.2f} | {np.round(importe_base[i], 2):12.2f} | {np.round(monto_iva[i], 2):9.2f} | {np.round(monto_beneficio[i], 2):15.2f} | {np.round(monto_reposicion[i], 2):16.2f} |")

print("---------------------------------------------------------------------------------------")
print(f"| TOTAL | {total_venta:11.2f} | {np.round(np.sum(importe_base), 2):12.2f} | {total_iva:9.2f} | {total_beneficio:15.2f} | {total_reposicion:16.2f} |")
print("---------------------------------------------------------------------------------------")


print("\nResumen de Totales")
print("---------------------------------")
print(f"Total de Ventas:        ${total_venta:7.2f}")
print(f"Total Recaudado de IVA: ${total_iva:7.2f}")
print(f"Total Beneficio Neto:   ${total_beneficio:7.2f}")
print(f"Total Fondo Reposición: ${total_reposicion:7.2f}")
print("---------------------------------\n")
# Comprobación (debe dar el Total Venta)
comprobacion = np.round(total_iva + total_beneficio + total_reposicion, 2)
print(f"Comprobación (IVA + Beneficio + Reposición): ${comprobacion:7.2f}")
````

7. Repetir el ejercicio anterior utilizando ufuncs.

````python
import numpy as np

# Datos de entrada
importes_venta = np.array([121.00, 50.00, 242.00, 80.00])
FACTOR_IVA_TOTAL = 1.21
FACTOR_BENEFICIO = 0.10
FACTOR_REPOSICION = 0.90

# Importe Base (sin IVA)
importe_base = np.divide(importes_venta, FACTOR_IVA_TOTAL)

# Monto de IVA
monto_iva = np.subtract(importes_venta, importe_base)

# Monto de Beneficio (10% del Base)
monto_beneficio = np.multiply(importe_base, FACTOR_BENEFICIO)

# Monto de Reposición (90% del Base)
monto_reposicion = np.multiply(importe_base, FACTOR_REPOSICION)

# Totales usando ufuncs
total_venta = np.sum(importes_venta)
total_iva = np.sum(monto_iva)
total_beneficio = np.sum(monto_beneficio)
total_reposicion = np.sum(monto_reposicion)

NUM_PRODUCTOS = importes_venta.size

print("---------------------------------------------------------------------------------------")
print("| Prod. | Venta Total | Importe Base | IVA (21%) | Beneficio (10%) | Reposición (90%) |")
print("---------------------------------------------------------------------------------------")

for i in range(NUM_PRODUCTOS):
    print(f"| {i+1:5} | {importes_venta[i]:11.2f} | {np.round(importe_base[i], 2):12.2f} | {np.round(monto_iva[i], 2):9.2f} | {np.round(monto_beneficio[i], 2):15.2f} | {np.round(monto_reposicion[i], 2):16.2f} |")

print("---------------------------------------------------------------------------------------")
print(f"| TOTAL | {total_venta:11.2f} | {np.round(np.sum(importe_base), 2):12.2f} | {total_iva:9.2f} | {total_beneficio:15.2f} | {total_reposicion:16.2f} |")
print("---------------------------------------------------------------------------------------\n")

print("Resumen de Totales")
print("---------------------------------")
print(f"Total de Ventas:        ${total_venta:7.2f}")
print(f"Total Recaudado de IVA: ${total_iva:7.2f}")
print(f"Total Beneficio Neto:   ${total_beneficio:7.2f}")
print(f"Total Fondo Reposición: ${total_reposicion:7.2f}")
print("---------------------------------\n")

comprobacion = np.round(np.add(np.add(total_iva, total_beneficio), total_reposicion), 2
print(f"Comprobación (IVA + Beneficio + Reposición): ${comprobacion:7.2f}")
````

11. En una ronda de una competición de petanca, cada participante debe enfrentarse con un
rival de forma aleatoria. A cada participante se le asigna un dorsal con un número según el
orden de llegada (1, 2, 3...) Solicitar el número de competidores y mostrar quién se enfrenta
con quién. En caso de que el número de competidores sea impar, se elegirá uno al azar que
pasará automáticamente a la siguiente ronda.

````python
import numpy as np

N = 5

dorsales = np.arange(1, N + 1)
print(f"Dorsales Iniciales: {dorsales}")

np.random.shuffle(dorsales)
print(f"Dorsales Barajados: {dorsales}")

if N % 2 != 0:

    pase_automatico = np.random.choice(dorsales)
    competidores_ronda = dorsales[dorsales != pase_automatico]

    print(f"\nPase Automático (Impar): Dorsal {pase_automatico}")
    print(f"Competidores para la ronda: {competidores_ronda}")

else:

    competidores_ronda = dorsales
    print("\nTodos los competidores participan en la ronda (Par).")


print("\nResultados del Emparejamiento")

num_a_emparejar = len(competidores_ronda)

for i in range(0, num_a_emparejar, 2):
    jugador_1 = competidores_ronda[i]
    jugador_2 = competidores_ronda[i + 1]

    partida_num = (i // 2) + 1

    print(f"Partida {partida_num}: Dorsal {jugador_1} VS Dorsal {jugador_2}")
````

12. El juego del bingo consiste en que los jugadores eligen una serie de números. A continuación,
se sacan bolas de un bombo mientras los jugadores tachan los números acertados. El primer
jugador que tacha todos sus números gana.
Implementar el juego del bingo para dos jugadores. A cada jugador se le asigna de forma
aleatoria 10 número comprendidos entre el 1 y el 90. Después se extraen números del bombo.
El programa debe indicar qué jugador gana y cuántos números le quedaba al otro.

````python
import numpy as np

MIN_NUMERO = 1
MAX_NUMERO = 90
NUM_CARTON = 10

carton_jugador_1 = np.sort(np.random.choice(range(MIN_NUMERO, MAX_NUMERO + 1), size=NUM_CARTON, replace=False))
carton_jugador_2 = np.sort(np.random.choice(range(MIN_NUMERO, MAX_NUMERO + 1), size=NUM_CARTON, replace=False))

print("JUEGO DE BINGO")
print(f"JUGADOR 1: {carton_jugador_1}")
print(f"JUGADOR 2: {carton_jugador_2}")
print("-" * 35)

bom_total = np.arange(MIN_NUMERO, MAX_NUMERO + 1)
np.random.shuffle(bom_total)

numeros_extraidos = np.array([])
jugador_ganador = None
i = 0

while i < len(bom_total) and jugador_ganador is None:

    bola_actual = bom_total[i]
    numeros_extraidos = np.append(numeros_extraidos, bola_actual)

    # Números restantes para el Jugador 1
    quedan_j1 = carton_jugador_1[~np.isin(carton_jugador_1, numeros_extraidos)]
    restantes_j1 = len(quedan_j1)

    # Números restantes para el Jugador 2
    quedan_j2 = carton_jugador_2[~np.isin(carton_jugador_2, numeros_extraidos)]
    restantes_j2 = len(quedan_j2)

    # Comprobar si hay ganador
    if restantes_j1 == 0:
        jugador_ganador = 1
    elif restantes_j2 == 0:
        jugador_ganador = 2

    # Avance de la simulación
    i += 1


print(f"Números extraídos: {numeros_extraidos}")
print(f"Última bola: {bola_actual}")
print("-" * 35)

if jugador_ganador:
    if jugador_ganador == 1:
        restantes_perdedor = restantes_j2
        jugador_perdedor = 2
    else:
        restantes_perdedor = restantes_j1
        jugador_perdedor = 1

    print(f"¡EL JUGADOR {jugador_ganador} HA GANADO!")
    print(f"Al Jugador {jugador_perdedor} le quedaban {restantes_perdedor} números.")
else:
    print("¡Empate! Se acabaron las bolas y nadie ganó.")
    print(f"Al J1 le quedan {restantes_j1} números. Al J2 le quedan {restantes_j2} números.")
````

13. La estatura media de los españoles (que es una población que tiene una distribución normal)

los valores máximos y mínimos. ¿Coinciden estos valores con la realidad? Comprobar que la
media es 174 cm.

````python
MEDIA_REAL = 174
DESV_ESTANDAR_REAL = 10
NUM_MUESTRAS = 100

print(f"Parámetros de la Población: Media={MEDIA_REAL} cm, Desv. Estándar={DESV_ESTANDAR_REAL} cm.")
print(f"Muestra aleatoria de {NUM_MUESTRAS} alturas.\n")
print("-" * 50)


alturas_generadas = np.round(np.random.normal(loc=MEDIA_REAL, scale=DESV_ESTANDAR_REAL, size=NUM_MUESTRAS),1 )

print(f"Muestra generada (primeros 10 valores):\n{alturas_generadas[:10]}...")
print("-" * 50)

altura_maxima = np.max(alturas_generadas)
altura_minima = np.min(alturas_generadas)
media_generada = np.mean(alturas_generadas)
desv_generada = np.std(alturas_generadas)

print("\nResultados del Análisis de la Muestra")
print(f"1. Valor Máximo (Muestra): {altura_maxima:.1f} cm")
print(f"2. Valor Mínimo (Muestra): {altura_minima:.1f} cm")

print("\nComprobación de la Media y Desviación")
print(f"Media Teórica (Población): {MEDIA_REAL} cm")
print(f"Media de la Muestra (np.mean): {media_generada:.3f} cm")
print(f"Desviación Estándar de la Muestra (np.std): {desv_generada:.3f} cm")

print("\nConclusión")

diferencia = abs(media_generada - MEDIA_REAL)

if diferencia < 1:
    print("La media de la muestra generada es muy similar a la media teórica (174 cm).")
    print(f"La diferencia es de solo {diferencia:.3f} cm.")
````

caso en el que el jugador gana.
Realizar un programa que mediante una distribución binomial simule cuántas veces ganaría
el jugador si realiza n tiradas. La distribución binomial, supone que solo existen dos casos
(éxito o fracaso), cada uno de ellos con una probabilidad de que ocurra: $p$, para el éxito y
$1-p$, para el fracaso.

````python
PROBABILIDAD_EXITO = 1 / 36  # p = 1/36 ≈ 0.02778
PROBABILIDAD_FRACASO = 1 - PROBABILIDAD_EXITO # 1 - p

try:
    N_TIRADAS = int(input("Introduce el número de tiradas (n) que deseas simular: "))
    if N_TIRADAS <= 0:
        print("El número de tiradas debe ser positivo.")
        exit()
except ValueError:
    print("Entrada inválida. Por favor, introduce un número entero.")
    exit()

resultados_simulacion = np.random.binomial(n=1, p=PROBABILIDAD_EXITO, size=N_TIRADAS)

veces_ganadas = np.sum(resultados_simulacion)

print("-" * 50)
print(f"Probabilidad de éxito (sacar 6-6): p = 1/36 ≈ {PROBABILIDAD_EXITO:.4f}")
print(f"Número total de tiradas simuladas (n): {N_TIRADAS}")
print("-" * 50)
print(f"El jugador ganó (sacó dos 6) un total de: {veces_ganadas} veces.")

valor_esperado_teorico = N_TIRADAS * PROBABILIDAD_EXITO
print(f"\nValor teórico esperado de ganancias (n * p): {valor_esperado_teorico:.2f} veces.")

if veces_ganadas > valor_esperado_teorico:
    print("Tu simulación obtuvo un poco más de victorias de lo esperado teóricamente. ¡Mucha suerte!")
elif veces_ganadas < valor_esperado_teorico:
    print("Tu simulación obtuvo un poco menos de victorias de lo esperado teóricamente.")
else:
    print("Tu simulación coincidió exactamente con el valor esperado teóricamente.")
````

La aplicación debe contar para n tiradas, cuantas veces gana (saca dos 6) y cuantas veces
empata (saca un único 6).

16. Un caminante que pasea con el algoritmo un pasito palante un pasito patrás, decide, de forma aleatoria en cada paso, dar un paso hacia adelante (1) o un paso hacia atrás (-1).
Calcular un paseo de 1000 pasos y mostrarlo gráficamente.

````python
import matplotlib.pyplot as plt

NUM_PASOS = 1000
PASOS_POSIBLES = np.array([1, -1]) # 1: Adelante, -1: Atrás

pasos_dados = np.random.choice(a=PASOS_POSIBLES, size=NUM_PASOS, p=[0.5, 0.5] )

posicion_inicial = np.array([0])

trayectoria = np.concatenate((posicion_inicial, np.cumsum(pasos_dados)))

posicion_final = trayectoria[-1]
max_posicion = np.max(trayectoria)
min_posicion = np.min(trayectoria)

print("Análisis del Paseo Aleatorio")
print(f"Número de pasos simulados: {NUM_PASOS}")
print(f"Posición Final (Distancia desde 0): {posicion_final}")
print(f"Máxima posición alcanzada: {max_posicion}")
print(f"Mínima posición alcanzada: {min_posicion}")
print("-" * 50)

indices_pasos = np.arange(NUM_PASOS + 1)

plt.figure(figsize=(12, 6))
plt.plot(indices_pasos, trayectoria, label='Posición del Caminante', color='blue')

plt.scatter(NUM_PASOS, posicion_final, color='red', s=50, zorder=5)
plt.text(NUM_PASOS, posicion_final + 1, f'Final: {posicion_final}', fontsize=10, color='red')

plt.title(f'(Un Pasito Palante/Patrás)', fontsize=14)
plt.xlabel('Número de Pasos', fontsize=12)
plt.ylabel('Posición (Distancia desde el Origen)', fontsize=12)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8, label='Origen (Posición 0)')
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.show()
````
