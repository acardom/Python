22. Eliminar todas las columnas salvo el nombre del pais y la capital. Implementar un juego que muestre un nombre de pais al azar y el usuario debe introducir su capital.

```python
dataframe_juego = dataframe[['nombre', 'capital']]

def juego_capitales(df):
    while True:

        random = np.random.randint(0, len(df))
        fila = df.iloc[random]

        pais = fila['nombre']
        capital_correcta = fila['capital']

        respuesta = input(f"¿Cuál es la capital de {pais}? (escribe 'salir' para terminar) \n").strip()

        if respuesta.lower() == 'salir':
            print("¡Gracias por jugar!")
            break

        if respuesta.lower() == capital_correcta.lower():
            print("¡Correcto!")
        else:
            print(f"Incorrecto. La capital de {pais} es {capital_correcta}.\n")

juego_capitales(dataframe_juego)
```


27. Crear una función que devuelva una diccionario de todos los paises cuyo nombre es compuesto (por ejemplo: «Antigua y Barbuda» 0 «Costa Ricas) junto al número de letras que tiene su nombre (no contar los espacios en blanco). Mostrar cuál de estos paises tiene el nombre más largo.



```python
def paises_compuestos(df):

    resultado = {}

    for pais in df['nombre']:
        if " " in pais:

            num_letras = len(pais.replace(" ", ""))
            resultado[pais] = num_letras

    return resultado

dic_paises = paises_compuestos(dataframe)
print(dic_paises)

pais_mas_largo = max(dic_paises, key=dic_paises.get)
longitud = dic_paises[pais_mas_largo]

print(f"\n El país con nombre más largo es '{pais_mas_largo}' con: \n {longitud} letras (sin espacios).")

```

31. A partir de una serie numérica aleatoria, construir un DataFrame con tres columnas: la propia serie, un valor que tenga una gran correlación y otra columna que tenga una correlación baja. Utilizar solo expresiones numéricas.

```python
x = np.random.randn(100)

y_high_corr = x * 3 + np.random.randn(100) * 0.1

y_low_corr = np.random.randn(100)

df = pd.DataFrame({
    "serie": x,
    "alta_correlacion": y_high_corr,
    "baja_correlacion": y_low_corr
})

print(df.head())
print("\nMatriz de correlación:\n", df.corr())
```

34. Crear una funcion de devuelva qué columnas de un dataset tiene algún valor nulo.
