# Series

Es un array unidireccional (1-D) etiquetado, puede contener cualquier tipom de dato

- Etiqueta 1 | Dato 1
- Etiqueta 2 | Dato 2
- Etiqueta 3 | ...

A estas etiquetas le llamaremos "indices"

Para crear una serie:

~~~
serie = pd.Series(Datos, indes=None)
~~~

Una serie se puede crear a partir de ciertos datos:
- Array-like
- dict-like
- escalar

index: array-like

```python
datos = ["Sevilla","Huelva","Málaga","Cádiz"]
ciudades = pd.Series(datos)
print(ciudades)
```

```python
etiquetas = ["s","h","m","c"]
ciudades = pd.Series(datos, index = etiquetas, name="Ciudades andaluzas")
print(ciudades)
```

```python
dic = {"pan": 1.2, "arroz": 2.3, "patatas": 3.2}
lista_compra = pd.Series(dic, index= ["pan", "arroz"])
print(lista_compra)
```

```python
etiquetas = ["a","b","c","d"]
poblacion = pd.Series(23, index=etiquetas)
print(poblacion)
```

# Dataframes

Un data frame puede crearse a partir de:
- Un diccionario:
  - 1-D ndarray
  - Listas
  - diccionarios
  - series

- una lista de:
  - listas
  - series

- un ndarray 2-D:
- Otro Dataframe

```python
# A partir de un diccionario de listas, sin idice:

pisos = {
    "m2":  [120, 85, 390],
    "precio": [185, 110, 450] # en miles de €
}

df_01 = pd.DataFrame(pisos)
print(df_01)
```

```python
# A partir de un diccionario de listas, con indices incluidos

pisos = {
    "m2":  [120, 85, 390],
    "precio": [185, 110, 450] # en miles de €
}

indices = ["piso1", "piso2", "piso3"]

df_02 = pd.DataFrame(pisos, index = indices)
print(df_02)
```

```python
# A partit de un ndarray 2-D

import numpy as np

datos = np.array([[1,2,3],[4,5,6]])

df_03 = pd.DataFrame(
    datos,
    columns = ["col1", "col2", "col3"],
    index = ["fil1", "fil2"]
    )

print(df_03)
```

# Funciones de datos
1. df.head(): muestra las 5 primeras filas
2. df.head(n): muesta las primeras n filas
3. df.tail(): muestra las ultimas 5 filas
4. df.tail(n): muestra las ultimas n filas
5. df.info(): muestra un resumen con iformación general sobre el DataFrame. esto incluye:
    - Numero total de filas y columnas
    - nombre y tipo de cada columna
    - cantidad de valores nulos por columna
    - uso de memoria
6. df.describe(): muestra estadisticas descriptivas del DataFrame, incluyendo:
    - count (cantidad de valores nulos)
    - mean (media)
    - std (desviaciones estandar)
    - min (minimo)
    - 25% / 50% / 75% (percentiles)
7. df.dtypes: muestra los tipos de datos de cada columna
8. df.index: muestra etiquetas de los indices de los Data Frame
9. df.columns: muestra los nombres de las columnas de DataFrame

# Crear columnas

Para crear columnas nuevas en un DataFrame ya existente:

```python
data_base = {
    "calorias": [20,30,50,4],
    "pulso": [100,105,120,85]
}

df_06 = pd.DataFrame(data_base)

df_06["nueva_col"] = df_06["calorias"] * df_06["pulso"]
display(df_06)
```

```python
data_base_02 = {
    "col1": [10,20,30],
    "col2": [40,50,60],
    "col3": [70,80,90]
}

df_07 = pd.DataFrame(data_base_02)
display(df_07)
print("")

df_07 = df_07.rename(
    columns = {
        "col1": "nuevo1",
        "col2": "nuevo2",
        "col3": "nuevo3"
        })

display(df_07)
```

# Eliminar contenido del DataFrame

~~~
df.drop(label = None, axis = 0, inplace = False)
~~~

- labels: Lista de etiquetas o columnas
- axis:
    - 0 (eje X) -> etiqueta/indices
    - 1 (eje Y) -> columnas

```python
# Eliminar columnas concretas
del df_06["pulso"]
display(df_06)
```

```python
serie = df_06.pop("nueva_col")
display(serie)
```

```python
datos = {
    "codigo": ["AP", "BL", "CG"],
    "alumno": ["Antonio Pérez", "Bea López", "Cristina Garcia"],
    "nota": [8,7,9]
}

df_08 = pd.DataFrame(datos)
display(df_08)

# Establezcamos la acolumna codigo como indice
df_08_nuevo = df_08.set_index("codigo", drop = True, inplace = False)
display(df_08_nuevo)
# drop = True va a eliminar la columna "codigo" despues de establecer como indice (si fuera False, la mantendria)
# inplace = False crea una nueva copia del DataFrame sin modificar el original (si fuese True, el DataFrame original se modificaria)

print("\n DataFrame con 'codigo' como indice:")
display(df_08)
```

# Otro ejemplo

# Filtros


```python
datos_ejemplo = {
    "nombre": ["Adrian Pavón", "Pablo González", "Alberto Cárdeno"],
    "edad": [21,19,22],
    "ciudad": ["El Garrobo", "Tocina y los Rosales", "Cumbres Mayores"]
}

df_09 =pd.DataFrame(datos_ejemplo)

filtro = df_09["edad"] < 20 # duevuelve una serie booleana
df_filtrado = df_09[filtro] # aplica el filtro del DataFrame

display(df_filtrado)
```

```python
# Filtrado de varias columnas

filtro1 = df_09["edad"] > 20
filtro2 = df_09["ciudad"] == "El Garrobo"

df_filtrado = df_09[filtro1 & filtro2]

display(df_filtrado)
```

```python
df_filtrado = df_09[filtro1 | filtro2]
display(df_filtrado)
```

```python
# Veamos como aplicar filtros en ua lista:

valores = ["El Garrobo", "Cádiz"]
filtro = df_09["ciudad"].isin(valores)

df_filtrado = df_09[filtro]

display(df_filtrado)
```

```python
# Como filtrar valores conocidos (no nulos) o nulos:

# Crear un DataFrame con valores nulos:
datos_ejemplo = {
    "nombre": ["Antonio", "Alberto", None, "Lucas"],
    "edad": [21, 22, 22, None]
}

df_10 = pd.DataFrame(datos_ejemplo)


# Filtro donde nombre no sea nulo
df_no_nulo = df_10[df_10["nombre"].notna()]

# Filtro donde filas "edad" sea nulo
df_nulo = df_10[df_10["edad"].isna()]

print("\n DataFrame con nombres no nulos:")
display(df_no_nulo)

print("\n DataFrame con edades nulas:")
display(df_nulo)
```

RECUERDA: Para filtrar con valores exactos usariamos:

~~~
df[df["columna"] == valor]
~~~

```python
if (df_10["edad"] == 21).any(): # verificara si hay al menos un True
    print("Hay personas de 21 años")
```

```python
# Podriamos usar un .all() para verificar si todas las filas cumplen una condicion
if (df_10["edad"] > 21).all():
  print("Todos son mayores de edad y tienen mas de 21 años")
else:
  print("No todos son mayores de 21")
```
